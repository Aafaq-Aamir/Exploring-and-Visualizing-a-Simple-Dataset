```python
# Task 1: Exploring and Visualizing the Iris Dataset

## Problem Statement
"""The goal of this task is to load, explore, and visualize the Iris dataset in order to understand
its structure, feature distributions, and relationships between variables. This will help build
a foundation in data exploration using Python, pandas, matplotlib, and seaborn.

## Objectives
- Load the Iris dataset using pandas.
- Inspect the dataset (shape, columns, first few rows).
- Compute summary statistics using `info()` and `describe()`.
- Visualize:
  - Scatter plots to show relationships between features.
  - Histograms to show feature value distributions.
  - Box plots to detect potential outliers.

This task focuses on basic EDA (Exploratory Data Analysis) and data visualization."""
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot style
sns.set(style="whitegrid")

# Load iris dataset from seaborn
iris = sns.load_dataset("iris")

iris.head()
```

```python
print("Shape of dataset:", iris.shape)
print("Column names:", iris.columns.tolist())
iris.info()
iris.describe()
```

```python
sns.pairplot(iris, hue="species", diag_kind="hist")
plt.suptitle("Iris Dataset Pairplot by Species", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species",
    palette="viridis"
)
plt.title("Sepal Length vs Petal Length by Species")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
for i, col in enumerate(["sepal_length", "sepal_width", "petal_length", "petal_width"], 1):
    plt.subplot(2, 2, i)
    sns.histplot(iris[col], kde=True)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris[["sepal_length", "sepal_width", "petal_length", "petal_width"]])
plt.title("Boxplots of Iris Features")
plt.show()
```

```python
## Conclusion and Insights
"""
- The dataset contains 150 samples with 4 numerical features and 1 target column (`species`).
- No missing values are present in this version of the dataset.
- Petal-related features (`petal_length` and `petal_width`) separate the three species much more
  clearly than sepal features, as seen in the pairplot and scatter plots.
- Histograms show that:
  - `setosa` petal lengths and widths are clearly smaller than the other species.
  - `versicolor` and `virginica` partially overlap but still show differences in petal size.
- Box plots suggest that there are no extreme outliers in the main features.

These observations indicate that the Iris dataset is clean, well-structured, and suitable for
classification tasks based on its numerical features."""
```
